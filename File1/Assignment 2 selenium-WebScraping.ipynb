```python
Q3: Scrape 100 reviews data from flipkart.com for iphone11 phone. You have to go the link:

https://www.flipkart.com/apple-iphone-11-black-64-gb/product-
reviews/itm4e5041ba101fd?pid=MOBFWQ6BXGJCEYNY&lid=LSTMOBFWQ6BXGJCEYNYZXSHRJ&marketplace=FLIPKART

As shown in the above page you have to scrape the tick marked attributes. These are:
1. Rating
2. Review summary
3. Full review
4. You have to scrape this data for first 100reviews.
Note: All the stepsrequired during scraping should be done through code only and not manually.
```

```python
driver = webdriver.Chrome()


#List for Ratings, summary, Full review

reviews = []
summary = []
ratings = []

#scraping the data

start=1
end=20
for page in range(start,end):
    link = "https://www.flipkart.com/apple-iphone-11-black-64-gb/product-reviews/itm4e5041ba101fd?pid=MOBFWQ6BXGJCEYNY&lid=LSTMOBFWQ6BXGJCEYNYZXSHRJ&marketplace=FLIPKART&page=" + str(page)
    driver.get(link)
    rating_tags=driver.find_elements(By.XPATH,'//div[@class="XQDdHH Ga3i8K"]') # scraping the ratings data
    for i in rating_tags:
        if(len(ratings)<100):
            get_rating=i.text
            ratings.append(get_rating)
        
    review_tags=driver.find_elements(By.XPATH,'//div[@class="ZmyHeo"]/div/div') # scraping the reviews data
    for i in review_tags:
        if(len(reviews)<100):
            get_review=i.text
            reviews.append(get_review)
        
    summary_tags=driver.find_elements(By.XPATH,'//div[@class="ZmyHeo"]/div/div') # scraping the summary data
    for i in summary_tags:
        if(len(summary)<100):
            get_summary=i.text
            summary.append(get_summary) 
            
    time.sleep(3)
    
```

```python
print(len(ratings), len(reviews), len(summary))
```

```python
Q4: Scrape data forfirst 100 sneakers you find whenyouvisitflipkart.com and search for “sneakers” inthe search
field.
You have to scrape 3 attributes of each sneaker:
1. Brand
2. ProductDescription
3. Price
As shown in the below image, you have to scrape the above attributes.
```

```python
driver = webdriver.Chrome()
link = 'https://www.flipkart.com'
driver.get(link)

#List for brand, prod_desc, price

brand = []
prod_desc = []
price = []

search = driver.find_element(By.XPATH,'/html/body/div[1]/div/div[1]/div/div/div/div/div[1]/div/div[1]/div/div[1]/div[1]/header/div[1]/div[2]/form/div/div/input')
search.send_keys("sneakers")

btn = driver.find_element(By.XPATH,'/html/body/div[1]/div/div[1]/div/div/div/div/div[1]/div/div[1]/div/div[1]/div[1]/header/div[1]/div[2]/form/div/button')
btn.click()

#scraping the data

start=1
end=10
for page in range(start,end):
    link = 'https://www.flipkart.com/search?q=sneakers&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page=' + str(page)
    driver.get(link)
    brand_tags=driver.find_elements(By.CLASS_NAME,'syl9yP') # scraping the ratings data
    for i in brand_tags:
        if(len(brand)<100):
            get_brand=i.text
            brand.append(get_brand)
        
    prod_desc_tags=driver.find_elements(By.CLASS_NAME,'WKTcLC') # scraping the reviews data
    for i in prod_desc_tags:
        if(len(prod_desc)<100):
            get_prod_desc=i.text
            prod_desc.append(get_prod_desc)
        
    price_tags=driver.find_elements(By.CLASS_NAME,'Nx9bqj') # scraping the summary data
    for i in price_tags:
        if(len(price)<100):
            get_price=i.text
            price.append(get_price) 
            
    time.sleep(3)
    
```

```python
Q6: Write a python program to scrape data for Top 1000 Quotes of All Time.
The above task will be done in following steps:
1. First get the webpagehttps://www.azquotes.com/
2. Click on TopQuote
3. Than scrap a)Quote b) Author c) Type Of Quotes
```

```python
driver = webdriver.Chrome()
link = 'https://www.azquotes.com/'
driver.get(link)

#List for brand, prod_desc, price

quote = []
author= []
type_of_quote = []

top_quotes = driver.find_element(By.XPATH,'/html/body/div[1]/div[1]/div[1]/div/div[3]/ul/li[5]/a')
top_quotes.click()

#scraping the data

start=0
end=10
for page in range(start,end):
    link = 'https://www.azquotes.com/top_quotes.html?p=' + str(page)
    driver.get(link)
    author_tags=driver.find_elements(By.CLASS_NAME,'author') # scraping the ratings data
    for i in author_tags:
        if(len(author)<=1000):
            get_author=i.text
            author.append(get_author)
        
    quote_tags=driver.find_elements(By.XPATH,'//div[@class="wrap-block"]/p/a[2]') # scraping the reviews data
    for i in quote_tags:
        if(len(quote)<=1000):
            get_quote=i.text
            quote.append(get_quote)
        
    type_tags=driver.find_elements(By.XPATH,'//div[@class="author"]/a') # scraping the summary data
    for i in type_tags:
        if(len(type_of_quote)<=1000):
            get_type=i.text
            type_of_quote.append(get_type) 
            
    time.sleep(3)
    
```

```python
print(len(author), len(type_of_quote), len(quote))
```
