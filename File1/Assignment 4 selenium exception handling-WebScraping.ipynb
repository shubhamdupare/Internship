```python
Scrape the details of most viewed videos on YouTube from Wikipedia. Url
= https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos You need to find following details: A)
Rank
B) Name
C) Artist
D) Upload date
E) Views
```

```python
driver = webdriver.Chrome()
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")

name = []
artist = []
upload_date =[]
views =[]


try:
    count = 0
    name_data = driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[1]//a[1]')
    for i in name_data:
        count+=1 
        
        if count % 2==0 :
            continue
        else :
            name.append(i.text)    
                       
    
    artist_data = driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[2]')
    for a in artist_data:
        artist.append(a.text)
    
    upload_data = driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[4]')
    for u in upload_data:
        upload_date.append(u.text)
    
    views_data = driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[3]')
    for v in views_data:
        views.append(v.text)
    
    
except NoSuchElementException:
    name.append('-')
    
```

```python
artist
```

```python
df = pd.DataFrame({'Name':name, 'Artist':artist, 'Upload Date':upload_date, 'Views':views})
df
```

```python
# Initialize WebDriver and open the website
driver = webdriver.Chrome()
driver.get("https://github.com/")
driver.maximize_window()

try:
    # Click on the "Economy" button
    open = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, '/html/body/div[1]/div[1]/header/div/div[2]/div/nav/ul/li[3]/button')))
    open.click()
    
    # Click on the "India" link
    trending = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, '/html/body/div[1]/div[1]/header/div/div[2]/div/nav/ul/li[3]/div/div[3]/ul/li[2]/a')))
    trending.click()

    # Wait for a while to ensure the next element is ready
    time.sleep(5)
    
except NoSuchElementException as e:
    print(f"An error occurred: {e}")
    


```

```python
title = []
description = []
count = []
language = []
url = []


url_data = driver.find_elements(By.XPATH,'//h2[@class="h3 lh-condensed"]/a')
for u in url_data:
    url.append(u.get_attribute("href"))

for i in url:
    driver.get(i)
    time.sleep(5)
    
    try:
        title_data = driver.find_element(By.XPATH, '//strong[@class="mr-2 flex-self-stretch"]/a')
        title.append(title_data.text.strip())
    except NoSuchElementException:
        title.append('-')
        
    try:
        description_data = driver.find_element(By.XPATH, '//p[@class="f4 my-3"]')
        description.append(description_data.text.strip())
    except NoSuchElementException:
        description.append('-')
        
    try:
        count_data = driver.find_element(By.XPATH, '//div[@class="BorderGrid about-margin"]/div[5]/div/h2/a/span')
        count.append(count_data.text.strip())
    except NoSuchElementException:
        count.append('-')
        
    try:  
        language_data = driver.find_element(By.XPATH, '/html/body/div[1]/div[4]/div/main/turbo-frame/div/div/div/div[2]/div[2]/div/div[6]/div/ul/li[1]/a/span[1]')
        language.append(language_data.text.strip())
    except NoSuchElementException:
        language.append('-')
                       
        
```

```python
df = pd.DataFrame({'Title':title, 'Description':description, 'Language':language,'Count':count})
df
```

```python
# Initialize WebDriver and open the website
driver = webdriver.Chrome()
driver.get("https:/www.billboard.com/")
time.sleep(2)
driver.maximize_window()

try:
    # Click on the "charts" button
    charts = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, '/html/body/div[3]/header/div/div[2]/div/div/div[2]/div[2]/div/div/nav/ul/li[1]/a')))
    charts.click()
    
    # Click on the "Hot 100" link
    trending = WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, '/html/body/div[3]/main/div[2]/div[1]/div[1]/div/div/div[3]/a')))
    time.sleep(10)
    trending.click()

    # Wait for a while to ensure the next element is ready
    time.sleep(5)
    
except (NoSuchElementException, TimeoutException) as e:
    print(f"An error occurred: {e}")
    
```

```python
#Data Scrape

song_name = []
artist_name = []
last_week_rank = []
peak_rank = []
week = []

try:
    song_name_data = driver.find_elements(By.XPATH, '//div[@class="o-chart-results-list-row-container"]/ul/li[4]/ul/li[1]/h3')
    for i in song_name_data:
        song_name.append(i.text.strip())
except NoSuchElementException:
    song_name.append('-')

try:
    artist_name_data = driver.find_elements(By.XPATH, '//div[@class="o-chart-results-list-row-container"]/ul/li[4]/ul/li[1]/span')
    for i in artist_name_data:
        artist_name.append(i.text.strip())
except NoSuchElementException:
    artist_name.append('-')

try:
    last_week_rank_data = driver.find_elements(By.XPATH, '//div[@class="o-chart-results-list-row-container"]/ul/li[4]/ul/li[4]/span')
    for i in last_week_rank_data:
        last_week_rank.append(i.text.strip())
except NoSuchElementException:
    last_week_rank.append('-')

try:
    peak_rank_data = driver.find_elements(By.XPATH, '//div[@class="o-chart-results-list-row-container"]/ul/li[4]/ul/li[5]/span')
    for i in peak_rank_data:
        peak_rank.append(i.text.strip())
except NoSuchElementException:
    peak_rank.append('-')

try:
    week_data = driver.find_elements(By.XPATH, '//div[@class="o-chart-results-list-row-container"]/ul/li[4]/ul/li[6]/span')
    for i in week_data:
        week.append(i.text.strip())
except NoSuchElementException:
    week.append('-')
```

```python
8. Details of Datasets from UCI machine learning repositories.
Url = https://archive.ics.uci.edu/ You
have to find the following details:
A) Dataset name
B) Data type
C) Task
D) Attribute type
E) No of instances
F) No of attribute G) Year
 Note: - from the home page you have to go to the Show All Dataset page through code. 
```

```python
# Initialize WebDriver and open the website
driver = webdriver.Chrome()
driver.get("https://archive.ics.uci.edu/")
time.sleep(2)
driver.maximize_window()

try:
    # Click on the "Viwe Datasets" button
    view_btn = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, '/html/body/div/div[1]/div[1]/main/div/div[1]/div/div/div/a[1]')))
    view_btn.click()
    
    # Wait for a while to ensure the next element is ready
    time.sleep(5)
    
except (NoSuchElementException, TimeoutException) as e:
    print(f"An error occurred: {e}")
    view_btn.click()
```

```python
#scrape data 

dataset_name = []
data_type = []
task = []
attribute_type = []
no_of_instances = []
no_of_attributes = []
year = []

# Adjust the range for the number of pages you want to scrape
start_page = 0
end_page = 100

accept_btn = WebDriverWait(driver, 40).until(EC.element_to_be_clickable((By.XPATH, '/html/body/div/div[1]/div[1]/div/div[2]/button')))
accept_btn.click()

# Loop to iterate over the pages
for page in range(start_page, end_page):
    
    time.sleep(5)  # Wait for the page to load properly
    
    try:
        btn_clk = WebDriverWait(driver, 40).until(EC.element_to_be_clickable((By.XPATH, '/html/body/div/div[1]/div[1]/main/div/div[2]/div[1]/div/label[2]')))
        btn_clk.click()
    except ElementClickInterceptedException:
        time.sleep(5)
        btn_clk.click()
    
    try:
        dataset_name_data = driver.find_elements(By.XPATH, '//h2[@class="truncate text-primary"]/a[1]')
        for element in dataset_name_data:
            dataset_name.append(element.text.strip())
    except NoSuchElementException:
        dataset_name.append('-')

    try:
        data_type_data = driver.find_elements(By.XPATH, '//div[@class="rounded-box grid grid-cols-8 gap-2 p-2 hover:bg-base-200"]/div[2]/div/div[2]/span')
        for element in data_type_data:
            data_type.append(element.text.strip())
    except NoSuchElementException:
        data_type.append('-')

    try:
        task_data = driver.find_elements(By.XPATH, '//div[@class="rounded-box grid grid-cols-8 gap-2 p-2 hover:bg-base-200"]/div[2]/div/div[1]/span')
        for element in task_data:
            task.append(element.text.strip())
    except NoSuchElementException:
        task.append('-')

    try:
        attribute_type_data = driver.find_elements(By.XPATH, '//tbody[@class="border"]/tr[1]/td[1]')
        for element in attribute_type_data:
            attribute_type.append(element.text.strip())
    except NoSuchElementException:
        attribute_type.append('-')

    try:
        no_of_instances_data = driver.find_elements(By.XPATH, '//div[@class="my-2 hidden gap-4 md:grid grid-cols-12"]/div[3]/span')
        for element in no_of_instances_data:
            no_of_instances.append(element.text.strip())
    except NoSuchElementException:
        no_of_instances.append('-')

    try:
        no_of_attributes_data = driver.find_elements(By.XPATH, '//div[@class="my-2 hidden gap-4 md:grid grid-cols-12"]/div[3]/span')
        for element in no_of_attributes_data:
            no_of_attributes.append(element.text.strip())
    except NoSuchElementException:
        no_of_attributes.append('-')

    try:
        year_data = driver.find_elements(By.XPATH, '//tbody[@class="border"]/tr[1]/td[3]')
        for element in year_data:
            year.append(element.text.strip())
    except NoSuchElementException:
        year.append('-')

    # Click the next button to go to the next page
    try:
        next_btn = WebDriverWait(driver, 50).until(EC.element_to_be_clickable((By.XPATH, '/html/body/div/div[1]/div[1]/main/div/div[2]/div[3]/div/button[2]')))
        time.sleep(2)
        next_btn.click()
        time.sleep(5)
    except ElementClickInterceptedException:
        print('Exception handling in process')
        next_btn.click()
```
