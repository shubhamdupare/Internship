```python
Q1) Write a python program which searches all the product under a particular product from www.amazon.in. The
product to be searched will be taken as input from user. For e.g. If user input is ‘guitar’. Then search for
guitars.
```

```python
#Q1 solution
def product_name():
    name=input("Enter the product name")
    return name

driver = webdriver.Chrome()
driver.get("https://www.amazon.in")

search=driver.find_element(By.XPATH, "/html/body/div[1]/header/div/div[1]/div[2]/div/form/div[2]/div[1]/input")
search.send_keys(product_name())

submit_btn=driver.find_element(By.XPATH, "/html/body/div[1]/header/div/div[1]/div[2]/div/form/div[3]/div/span/input")
submit_btn.click()
    
time.sleep(3)
```

```python
Q2) In the above question, now scrape the following details of each product listed in first 3 pages of your search
results and save it in a data frame and csv. In case if any product has less than 3 pages in search results then
scrape all the products available under that product name. Details to be scraped are: "Brand
Name", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability" and
“Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“.
```

```python
# Scrape all product Urls

product_urls = []
start = 0
end = 2

for page in range(start,end): #loop for scrapping 3 pages
    url=driver.find_elements(By.XPATH,'//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')
    for i in url:
        product_urls.append(i.get_attribute("href"))   #appending the urls
    next_btn=driver.find_element(By.XPATH,'//a[@class="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator"]')   
    next_btn.click()
    time.sleep(3)
```

```python
len(product_urls)
```

```python
brand =[]
name = []
price =[]
return_exc =[]
avail =[]
delivery =[]


for url in product_urls:
    driver.get(url)
    time.sleep(2)
    
    try:
        brand_tag=driver.find_element(By.XPATH,'//span[@class="a-size-small tabular-buybox-text-message"]/a')
        brand.append(brand_tag.text)
    except NoSuchElementException:
        brand.append('-')
        
    try:
        name_tag=driver.find_element(By.XPATH,'//span[@class="a-size-large product-title-word-break"]')
        name.append(name_tag.text)
    except NoSuchElementException:
        name.append('-')
        
    try:
        price_tag=driver.find_element(By.XPATH,'//span[@class="a-price-whole"]')
        price.append(price_tag.text)
    except NoSuchElementException:
        price.append('-')
        
    try:
        return_exc_tag=driver.find_element(By.XPATH,'//div[@id="RETURNS_POLICY"]/span/div[2]/span')
        return_exc.append(return_exc_tag.text)
    except NoSuchElementException:
        brand.append('-')    
        
    try:
        avail_tag=driver.find_element(By.XPATH,'//span[@class="a-size-medium a-color-success"]')
        avail.append(avail_tag.text)
    except NoSuchElementException:
        avail.append('-')        
        
    try:
        delivery_tag=driver.find_element(By.XPATH,'//div[@class="a-spacing-base"]/span/span')
        delivery.append(delivery_tag.text)
    except NoSuchElementException:
        delivery.append('-')     
    
```

```python
df = pd.DataFrame({'Name' : name, 'Brand' : brand, 'Price' : price, 'Return/Exchange' : return_exc, 'Availiblity' : avail, 'Delivery' : delivery })
df
```

```python
Q3) Write a python program to access the search bar and search button on images.google.com and scrape 10
images each for keywords ‘fruits’, ‘cars’ and ‘Machine Learning’, ‘Guitar’, ‘Cakes’.
```

```python
driver = webdriver.Chrome()

keywords = ['fruits', 'cars', 'Machine Learning', 'Guitar', 'Cakes']
img_url = []
```

```python

# Search keyword wise 10 images 
for keyword in keywords:
    # Navigate to Google Images
    driver.get("https://www.google.com/imghp?hl=en&authuser=0&ogbl")
    
    
    # Find the search bar and enter the keyword
    search_bar = driver.find_element(By.XPATH, '//textarea[@class="gLFyf"]')
    search_bar.send_keys(keyword)
    
    # Press Enter to perform the search
    search_btn = driver.find_element(By.XPATH, '/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/button')
    search_btn.click()
    
    # Wait for images to load
    time.sleep(3)
    
    # Scroll down to lodd more images
    for _ in range(20):
        driver.execute_script("window.scrollTo(0,1000)")
        time.sleep(2)
        

    images=driver.find_elements(By.XPATH,'//img[@class="YQ4gaf"]')
    for image in images:
        source=image.get_attribute('src')
        if source is not None:
            if(source[0:4] == 'http'):
                img_url.append(source)

    for i in range(len(img_url)):
        if i > 10:
            break
        print("Downloading {0} of {1} images" .format(i,10))
        response=requests.get(img_url[i])
        file = open(r"C:\Users\shubham\Desktop\Fliprobo"+str(i)+".jpg", "wb")
        file.write(response.content)
```

```python
Q6) Write a program to scrap all the available details of best gaming laptops from digit.in.

```

```python
driver = webdriver.Chrome() 

driver.get('https://www.digit.in/top-products/best-gaming-laptops-40.html')

time.sleep(3)  # Wait for 3 seconds

# Extract the Url of laptops 
laptop_url = []

laptop_tag = driver.find_elements(By.XPATH,'//div[@class="rh_gr_img_first"]/figure/a')

for i in laptop_tag:
    laptop_url.append(i.get_attribute('href'))
    
time.sleep(3)    
```

```python
laptop_url
```

```python

# extract deetails from the each URL

name=[]
model=[]
os=[]
display_size=[]
processor=[]
storage=[]
ram=[]

for i in laptop_url:
    driver.get(i)
    time.sleep(3)
    
    try:
        name_tag=driver.find_element(By.XPATH,'//h1[@class="floatleft tabletblockdisplay pr20 "]')
        name.append(name_tag.text)
    except NoSuchElementException:
        name.append('-')
    
    try:
        model_tag=driver.find_element(By.XPATH,'/html/body/div[1]/div[3]/div[1]/div/div/div/div/div[2]/div[4]/div[3]/div[2]/table/tbody/tr[2]/td/table/tbody/tr[3]/td')
        model.append(model_tag.text)
    except NoSuchElementException:
        model.append('-')
    
    try:
        os_tag=driver.find_element(By.XPATH,'//div[@class="value"]/p[2]/strong')
        os.append(os_tag.text)
    except NoSuchElementException:
        os.append('-')
    
    try:
        display_size_tag=driver.find_element(By.XPATH,'/html/body/div[1]/div[3]/div[1]/div/div/div/div/div[2]/div[1]/div[4]/div[3]/div/div/ul/li[2]/div/p[2]/strong')
        display_size.append(display_size_tag.text)
    except NoSuchElementException:
        display_size.append('-')
    
    try:
        processor_tag=driver.find_element(By.XPATH,'/html/body/div[1]/div[3]/div[1]/div/div/div/div/div[2]/div[1]/div[4]/div[3]/div/div/ul/li[3]/div/p[2]/strong')
        processor.append(processor_tag.text)
    except NoSuchElementException:
        processor.append('-')
        
    try:
        storage_tag=driver.find_element(By.XPATH,'/html/body/div[1]/div[3]/div[1]/div/div/div/div/div[2]/div[1]/div[4]/div[3]/div/div/ul/li[4]/div/p[2]/strong')
        storage.append(storage_tag.text)
    except NoSuchElementException:
        storage.append('-')
        
    try:
        ram_tag=driver.find_element(By.XPATH,'/html/body/div[1]/div[3]/div[1]/div/div/div/div/div[2]/div[4]/div[3]/div[2]/table/tbody/tr[8]/td/table/tbody/tr[1]/td/p')
        ram.append(ram_tag.text)
    except NoSuchElementException:
        ram.append('-')    
```

```python
df = pd.DataFrame({'Laptop Name':name,'Model Name':model,'Operating System':os,'Display':display_size,'Processor':processor,'Storage':storage,'Ram':ram})
df
```
